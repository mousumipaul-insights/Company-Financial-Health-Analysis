## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print('Libraries loaded ✅')
```

## 2. Load Raw Financial Data

```python
# Update path to your actual file
df = pd.read_excel('../data/raw/financial_statements.xlsx', sheet_name='Income Statement', index_col=0)
df.head()
```

## 3. Inspect Data

```python
print('Shape:', df.shape)
print('\nData Types:')
print(df.dtypes)
print('\nMissing Values:')
print(df.isnull().sum())
```

## 4. Clean & Standardize

```python
# Rename columns to standard format
df.columns = [str(c).strip() for c in df.columns]

# Fill or drop missing values
df = df.fillna(0)

# Convert to numeric where applicable
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.head()
```

## 5. Save Cleaned Data

```python
df.to_csv('../data/processed/cleaned_financials.csv')
print('Cleaned data saved ✅')
```
