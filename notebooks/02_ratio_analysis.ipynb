## 5. 5-Year Ratio Trend

```python
import matplotlib.pyplot as plt

years = ['FY2019', 'FY2020', 'FY2021', 'FY2022', 'FY2023']

# Example: Plot ROE trend (replace with actual computed values)
roe_trend = [12.1, 13.4, 11.8, 14.2, 15.6]

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(years, roe_trend, marker='o', color='steelblue', linewidth=2)
ax.set_title('Return on Equity (ROE) – 5 Year Trend')
ax.set_ylabel('ROE (%)')
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('../outputs/charts/roe_trend.png', dpi=150)
plt.show()
```
