## 1. Setup

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
sns.set_theme(style='whitegrid')
print('Setup complete ✅')
```

## 2. Revenue & Profit Trend

```python
years = ['FY2019','FY2020','FY2021','FY2022','FY2023']
revenue     = [28000, 29500, 27800, 33000, 37500]   # Replace with actual
net_income  = [2100,  2300,  1900,  2800,  3400]

fig, ax1 = plt.subplots(figsize=(9, 4))
ax2 = ax1.twinx()
ax1.bar(years, revenue, color='steelblue', alpha=0.7, label='Revenue')
ax2.plot(years, net_income, 'o-', color='darkorange', linewidth=2, label='Net Income')
ax1.set_ylabel('Revenue (₹ Cr)', color='steelblue')
ax2.set_ylabel('Net Income (₹ Cr)', color='darkorange')
ax1.set_title('Revenue vs Net Income – 5 Year Trend')
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.85))
plt.tight_layout()
plt.savefig('../outputs/charts/revenue_net_income.png', dpi=150)
plt.show()
```

## 3. Ratio Radar / Bar Chart

```python
ratios = {'Current Ratio':1.8, 'Quick Ratio':1.2, 'ROE (%)':15.6,
         'ROA (%)':8.2, 'Gross Margin (%)':42.1, 'Net Margin (%)':9.1,
         'D/E Ratio':0.45, 'Interest Coverage':8.3}

fig, ax = plt.subplots(figsize=(10, 5))
colors = ['#2196F3']*3 + ['#4CAF50']*3 + ['#F44336']*2
ax.barh(list(ratios.keys()), list(ratios.values()), color=colors)
ax.set_title('Key Financial Ratios – FY2023')
ax.set_xlabel('Value')
plt.tight_layout()
plt.savefig('../outputs/charts/ratio_bar.png', dpi=150)
plt.show()
```

## 4. Cash Flow Waterfall

```python
categories = ['Operating CF', 'Investing CF', 'Financing CF', 'Net Change']
values     = [5200, -2100, -1500, 1600]
colors     = ['green' if v >= 0 else 'red' for v in values]

fig, ax = plt.subplots(figsize=(7, 4))
ax.bar(categories, values, color=colors, edgecolor='white', width=0.5)
ax.axhline(0, color='black', linewidth=0.8)
ax.set_title('Cash Flow Summary – FY2023 (₹ Crores)')
ax.set_ylabel('₹ Crores')
for i, v in enumerate(values):
    ax.text(i, v + (150 if v >= 0 else -300), f'₹{v:,}', ha='center', fontsize=9)
plt.tight_layout()
plt.savefig('../outputs/charts/cashflow_waterfall.png', dpi=150)
plt.show()
```

## 5. DCF Intrinsic Value vs Market Price

```python
labels = ['Intrinsic Value (DCF)', 'Current Market Price']
values = [742.5, 680.0]   # Replace with actual
colors = ['#4CAF50', '#2196F3']

fig, ax = plt.subplots(figsize=(6, 4))
bars = ax.bar(labels, values, color=colors, width=0.4)
ax.set_title('DCF Intrinsic Value vs Market Price')
ax.set_ylabel('Price per Share (₹)')
for bar, val in zip(bars, values):
    ax.text(bar.get_x() + bar.get_width()/2, val + 5, f'₹{val}', ha='center', fontweight='bold')
plt.tight_layout()
plt.savefig('../outputs/charts/dcf_vs_market.png', dpi=150)
plt.show()
```
